```python
# THIS CODE COMBINES THE ERROR TABLE AND FEATURE TABLE (OUTER JOIN)

train_df = pd.read_csv('/Users/punit/SBU/Acads/DSF/Assignments/HW2/train_2016_v2.csv') # TODO: Need to change these files' locations

parcel_value_list = train_df['parcelid']

df_filter_tained_parcel_ids = df[df.parcelid.isin(parcel_value_list)]

# df_filter_tained_parcel_ids # 90150 rows
# len(df_filter_tained_parcel_ids.parcelid.unique())
parcel_value_list2 = df_filter_tained_parcel_ids['parcelid']
# len(parcel_value_list2)
train_df_2 = train_df[train_df.parcelid.isin(parcel_value_list2)]
sorted_df_filter_tained_parcel_ids = df_filter_tained_parcel_ids.sort_values(by='parcelid', ascending=1)

train_df_2 = train_df_2.drop_duplicates(['parcelid'], keep='last') #90150
sorted_train_df_2 = train_df_2.sort_values(by='parcelid', ascending=1)

train_df_joined = pd.merge(sorted_df_filter_tained_parcel_ids, sorted_train_df_2, on='parcelid', how='outer')



```

```python
# THIS IS JUST A REFENRECE CODE, NEED TO CHECK BEFORE REMOVING THE CELL!

import matplotlib.pyplot as plt
df_clean = train_df_joined[numpy.isfinite(train_df_joined['structuretaxvaluedollarcnt'])]
result = df_clean[numpy.isfinite(df_clean['logerror'])]
result = result [result ['structuretaxvaluedollarcnt'] < 2000000] # for better visulizations
# result = result [result ['finishedsquarefeet15'] < 25000] # Had to remove the outliers with zero bedrooms for better visulizations
result[['structuretaxvaluedollarcnt','logerror']].plot.scatter('structuretaxvaluedollarcnt', 'logerror')
```

```python
# THIS IS THE CODE FOR LINEAR REGRESSION WITH ONE FEATURE (only 80% training data, 20% test data)

import numpy as np
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import ensemble

train_df_joined_no_logerror_nan = train_df_joined[numpy.isfinite(train_df_joined['logerror'])]
train_df_joined_no_logerror_strctdollar_nan = train_df_joined_no_logerror_nan[numpy.isfinite(train_df_joined_no_logerror_nan['structuretaxvaluedollarcnt'])]

train_df_joined_no_logerror_strctdollar_nan.isnull().any()

train_x1_full = train_df_joined_no_logerror_strctdollar_nan.structuretaxvaluedollarcnt.values

slice_num = int(len(train_x1_full)*0.8)
train_x1 = train_x1_full[:slice_num]

train_x = train_x1.reshape(train_x1.shape[0], 1)

# np.isnan(np.min(train_x))


# # # tain_x2_unclean = train_df_joined.finishedsquarefeet15.values
# # # df_clean = train_df_joined[numpy.isfinite(train_df_joined['structuretaxvaluedollarcnt'])]

# # # train_df_joined_no_logerror_nan.isnull().sum()


train_y_full = train_df_joined_no_logerror_strctdollar_nan.logerror.values
train_y = train_y_full[:slice_num]

# Creating test data for the ML model
test_x = train_x1_full[slice_num:]
test_x = test_x.reshape(test_x.shape[0], 1)

test_y = train_y_full[slice_num:]

# train_x.shape

# # np.isnan(np.min(train_y))

# # # # tain_x = tain_x.reshape(length, 1)
# # # # train_y = train_y.reshape(length, 1)

# # Gradient boosting

# clf = ensemble.GradientBoostingRegressor(n_estimators=400, max_depth=5, min_samples_split=2, learning_rate = 0.1, loss = 'ls')
# clf.fit(train_x, train_y)
# clf.score(test_x, test_y)

# Linear Regression starts from here

regr = linear_model.LinearRegression()
regr.fit(train_x, train_y)

# # The coefficients
print('Coefficients: \n', regr.coef_)

pred_y = regr.predict(test_x)

# Score
print ("The score is : %.9f" % regr.score(test_x, test_y))

# # The mean squared error
print("Mean squared error: %.9f"
      % mean_squared_error(test_y, pred_y))

# # Explained variance score: 1 is perfect prediction
print('Variance score: %.9f' % r2_score(test_y, pred_y))


# plot it as in the example at http://scikit-learn.org/
plt.rcParams['figure.figsize']=(15,10)
plt.xlabel("Structure Value ($)")
plt.ylabel("Log error")
plt.scatter(train_x, train_y,  color='black')
plt.plot(test_x, pred_y, color='blue', linewidth=3)
plt.show()

```
