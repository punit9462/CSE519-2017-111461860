*~~~~ Building a scoring function to determine the desirability of the houses : *

```python
# Loading the properties file into a dataframe

import pandas as pd
import numpy as numpy
import scipy.spatial.distance as scp_dst
from scipy.spatial import distance
from sklearn.cluster import KMeans
import matplotlib.colors as clrs
import matplotlib.pyplot as plt
from sklearn import preprocessing

df = pd.read_csv('properties_2017.csv', low_memory=False)
```

```python
# Cleaning the data required to compute the scoring function

df_clean_for_scoring = df[numpy.isfinite(df['finishedsquarefeet15'])]
df_clean_for_scoring = df_clean_for_scoring[numpy.isfinite(df_clean_for_scoring['roomcnt']) & df_clean_for_scoring['roomcnt'] != 0]
df_clean_for_scoring = df_clean_for_scoring[numpy.isfinite(df_clean_for_scoring['structuretaxvaluedollarcnt'])]
df_clean_for_scoring = df_clean_for_scoring[numpy.isfinite(df_clean_for_scoring['yearbuilt'])]
df_clean_for_scoring.head(10) # has 396 rows (these might be the houses)
```

```python
# Compute the scoring function

df_clean_for_scoring ['costperarea'] = df_clean_for_scoring['structuretaxvaluedollarcnt'] / df_clean_for_scoring['finishedsquarefeet15']
df_clean_for_scoring ['desirabilityscore'] = (df_clean_for_scoring['roomcnt']*2) + (df_clean_for_scoring['yearbuilt']) - (df_clean_for_scoring['structuretaxvaluedollarcnt'] / df_clean_for_scoring['finishedsquarefeet15']) 
df_sorted_on_score = df_clean_for_scoring.sort_values(['desirabilityscore'], ascending=[False])
```

```python
# Top 10 desirable properties

df_sorted_on_score_limited_columns = df_sorted_on_score.loc[:,['parcelid','roomcnt','yearbuilt','costperarea','desirabilityscore']]
df_sorted_on_score_limited_columns.head(10)
```

```python
# 10 least desirable properties

df_sorted_on_score_limited_columns = df_sorted_on_score_limited_columns.sort_values(['desirabilityscore'], ascending=[True])
df_sorted_on_score_limited_columns.head(10)
```

Scoring function is a mathematical function that maps the feature variables with a real number that helps us in identifying the top and bottom records of the dataset. For Zillow challenge, we have roughly 60 input features for each property and it’s not practical to use all of these features and come up with a scoring function. We have to selectively consider the input features in such a way that (1) our scoring function makes sense and (2) we end up using less number of input features. For that we introduce the following definition of desirability,

For us, the most desirable house will have more number of rooms (roughly speaking, that is more area) and less cost per area. This way, we can get the best “value for money” property. Additionally, we also want to buy the recent houses. We don’t want to spend money on properties which are very old. Therefore, we are considering following features (columns) in the dataset in order to model the mentioned definition of desirability. 

- finishedsquarefeet15 (which is Total area of the property),
- roomcnt (number of rooms in the house),
- structuretaxvaluedollarcnt (assessed value of built structure on the parcel),
- yearbuilt (The Year the principal residence was built)

Now, since we want to consider more room counts and recent houses, our scoring function will have following term, 
- (2 x room_count) + year_built (Note that we are scaling the room count by 2 to give it more importance than the other features.)

We also don’t want to end up having big houses in the list which are very expensive and we can’t afford them. To reduce the possibility of that event, we are decreasing the score of the houses which have more price per area and hence, our scoring function will become as following, 

```python
from IPython.display import Latex
Latex(r"""\begin{eqnarray}
desirability\_score = (2*room\_count) + year\_built - \frac{structure\_tax\_value\_dollar} {total\_area} 
\end{eqnarray}""")
```

How well the scoring function worked?

- We calculated the desirability_score for the properties of Zillow challenge and following are the observations,  
    - It can be seen from the "Top 10 desirable properties" that the properties with very less cost per area and having more rooms are ranked higher. Also, note that they are also built very recently.  
    - We can see the opposite of this in the list titled "10 least desirable properties" - that is properties with less rooms and more value are scored relatively very low. 

This way, it seems that our scoring function did a pretty good job in ranking the properties based on the desired features.

Distance functions can generally be used to classify records in different categories. Similar records (having similar sort of features) should have same distance values from each other. At the same time, “different-looking” entities should be separated by relatively larger distance. In this task, we will calculate the distance between two Zillow properties. In particular, we are considering the Euclidean distance (which is the pairwise distance function) as following,

if p = (p1, p2,..., pn) and q = (q1, q2,..., qn) are two data points in the dataset, then the distance (d) from p to q, or from q to p is given by the following Pythagorean formula,

```python
from IPython.display import Latex
Latex(r"""\begin{eqnarray}
d(p,q) = \sqrt{(q_1-p_1)^2 + (q_2-p_2)^2 + (q_3-p3)^2 + .... + (q_n-p_n)^2}
\end{eqnarray}""")
```

where,
- all pi’s and qi’s are various feature values. Note that here if the two points p and q are same, the distance value will be zero, which is one of the desired properties for our distance function for Zillow houses and more dissimilar houses will have larger errors and will lead to larger distance values. Now, we will discuss which feature vectors we should consider for distance vector. 

For housing properties, the following assumptions can be made, 

    1. Two housing properties will mostly be similar if they are geographically close to each other.
    2. If the number of rooms, total area and number of units (duplex, triplex etc) are the same for two houses, 
    they may be considered same.

In the Zillow dataset, we have following features to calculate the disntance as explained above, 

    1. Geographic variables : latitude, longitude and regionidzip (zipcode of the place)
    2. Property-specific variables: finishedsquarefeet15 (Total area), unitcnt (Number of units the structure is 
    built into), roomcnt (Total number of rooms in the principal residence)

This way, our distance function has five features :- latitude, longitude, regionidzip, finishedsquarefeet15, unitcnt, roomcnt.

*Some interesting points to note*

-	To have a uniform effect of various feature columns across all the datapoints, we have normalized our features between 0 and 1 – which is explained as a python comment in the notebook. If we don’t normalize, the function will incorrectly weigh the larger values more and our distance may not turn out to be optimal. 

-	After normalization, we calculated the distance values of some property-pairs and we found the number to be really small to compare. Hence, we are scaling up the normal Euclidean distance that we get from the above formula by 10^6 so that we get a distance number that is easier to work with and visualize.

*~~~~ Incorporate one external data set into the model*

- We considered the mediansalary of the areas corresponding to the properties. We found a dataset online (external_zipcode_mediansalary_mapping.csv) where we have zipcode to median salary mapping.


- Since the dataset released by Zillow does not have the correct zipcodes, we had to map the regionzip (fake zipcodes of Zillow dataset) to the actual zipcode. For that, we used another external dataset which had zipcode to lat/long mapping and we used "least-distance" creteria to assign the correct zipcode to the properties based on their lat/long. This way we generated the parcel_id to correct zipcode mapping (external_parcelid_correct_zipcode_mapping.csv). Following is the python script used to club the data. Notice that it uses the least square distance method to get the correct zipcode given the lat/long of the property -- line: distance = math.sqrt(((lat-latRef)*(lat-latRef)) + ((lng-lngRef)*(lng-lngRef)))


```python
# Script used to generate the parcel_id to correct zipcode mapping

import requests
import csv
from pygeocoder import Geocoder
import math

f = open('out_zipcode.csv','a')
f.write('parcelid,zipcode' + '\n')

with open('properties_2017.csv') as csvfile:
        readCSV = csv.reader(csvfile, delimiter=',')
        dates = []
        colors = []
        x = 0;
        cnt = 0;
        for row in readCSV:
            if(x==0): 
                x = x+1
                continue
            pid = row[0]
            lat = float(row[24])/1000000
            lng = float(row[25])/1000000
            with open('zipcode_data.csv') as csvfile2:
                readCSV2 = csv.reader(csvfile2, delimiter=',')
                y = 0;
                min_val = 2147483647
                zip_code = 0
                for row2 in readCSV2:
                    if(y==0): 
                        y = y+1
                        continue
                    latRef = float(row2[1])
                    lngRef = float(row2[2])
                    distance = math.sqrt(((lat-latRef)*(lat-latRef)) + ((lng-lngRef)*(lng-lngRef)))
                    if(distance < min_val) :
                        min_val = distance
                        zip_code = row2[0]
                out_text = pid + ',' + zip_code
                cnt = cnt + 1
                print ("Count: " + str(cnt) + " Parcel " + pid + " done")
                f.write(out_text+'\n')
f.close()
```

*~~~~ Build the best prediction model to solve the Zillow task*

- In this task, we will attempt to improve our model to predict the logerror on Zillow data. At first we will discuss the data cleaning and preparation methods that we used then we will discuss the actual model and training and finally we will present the results of our model and it's evaluation.


- Zillow data has 57 features for a given property. We want to select the best features among these to consider for our models. One of the ways we can do this is by using feature selection algorithms. There are various selection algorithms that can be used. We can eliminate the features which has very low variance, as low variance means that this feature doesn't convey a lot of information. And feature has similar values for most of the data points. Another way we can eliminate the features is using univariate feature selection algorithms which uses univariate statistical tests such as X^2. 


- In this experiment, however, we are going to use sklearn's SelectFromModel feature selection algorithm. It uses an estimator and selects the features having weights or coefficients only above a specified threshold. Thresholds can be specified numerically or they can be determined heuristically by sklearn library. We are going to use Lasso regression for selecting the features. But before we could do that, we have to prepare our data to be used for regression. 


- For preparing the data, we will be keeping things very simple - We will first encode all the categorical features in the data and eliminate id field from the data.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data=pd.read_csv("properties_2016 2.csv",low_memory=False)
logerrordata=pd.read_csv('train_2016_v2.csv',low_memory=False)
logerrordata=pd.merge(data,logerrordata,on='parcelid')
```

```python
X=logerrordata.drop('parcelid',1)
Y=X['logerror']
X=X.drop('logerror',1)
```

```python
# Function that encodes all the categorical features in dataframe and returns it back
from sklearn import preprocessing
def EncodeCatecoricalVariables(X):
    X_cat = X.select_dtypes(include=[object])
    le = preprocessing.LabelEncoder()
    X_cat=X_cat.apply(le.fit_transform)
    X=X.drop(X_cat.columns,1)
    X=pd.concat([X,X_cat],axis=1)
    return X
```

```python
X=EncodeCatecoricalVariables(X)
X.head()
```

- Now, we have encoded all the categorical variables. We still have to deal with missing values before we can proceed further with feature selection. We will keep this simple and replace all the missing values with mean value of feature.

```python
X=X.fillna(X.mean())
```

- We are almost ready to apply feature selection algorithms to our data. We will be using Lasso regression for our feature selection. It is recommended that our features be normalized to get best results out of lasso regression. So let's normalize our data.

```python
from sklearn import preprocessing
def Normalize(X):
    X_normalizd=preprocessing.Normalizer().fit_transform(X) 
    X_normalizd=pd.DataFrame(X_normalizd)
    X_normalizd.columns=X.columns
    X_normalizd.index=X.index
    return X_normalizd
```

```python
X_normalized=Normalize(X)
```

- Now we are ready to apply our feature selection algorithm.

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LassoCV
clf = LassoCV()

# Set a minimum threshold of 0.1
sfm = SelectFromModel(clf, threshold=0.1)
sfm.fit(X_normalized, Y)
i=1
while i <10:
    sfm.threshold += 0.1
    X_transform = sfm.transform(X_normalized)
    i+=1
feature_index=sfm.get_support(indices=True)
print(feature_index)
```

- We got 17 features by applying the feature selection algorithm. Now let's use these features to train a simple linear regression model.

```python
X_feature_selected=X_normalized.iloc[:,feature_index]
X_feature_selected.head()
```

```python
from sklearn.linear_model import LinearRegression
import sklearn.cross_validation
X_train,X_test,Y_train,Y_test=sklearn.cross_validation.train_test_split(X_feature_selected,Y,test_size=0.25,random_state=5)
linearmodel=LinearRegression()
linearmodel.fit(X_train,Y_train)
train_logerror=linearmodel.predict(X_train)
test_logerror=linearmodel.predict(X_test)

mse_train=np.mean((Y_train-train_logerror)**2)
mse_test=np.mean((Y_test-test_logerror)**2)
print("training_error",mse_train)
print("testing_error",mse_test)
r2=linearmodel.score(X_test,Y_test)
print("r2",r2)
coefDF=pd.DataFrame(zip(X_train.columns,linearmodel.coef_),columns=['features','Coefficients'])
coefDF
```

- The linear model is getting mean squared error 0.0264 on training data and 0.024 on test data. The r^2 value seems low for our linear model. Now let's look at residual graph to see how model actually performed.

```python
import matplotlib.pyplot as plt
fig = plt.figure()
fig.set_size_inches(18.5, 10.5)
plt.scatter(train_logerror,train_logerror-Y_train,s=1,marker='.',alpha=1)
plt.title('Residuals for train data')
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
fig.set_size_inches(15.5, 8.5)
plt.scatter(test_logerror,test_logerror-Y_test,s=1,marker='.',alpha=1)
plt.title('Residuals for test data')
plt.show()
```

- As seen from residual graphs we can see that linear model works well for most of the data. But it fails badly for some data points. Now let's see what permutation test tell us about our model.

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import permutation_test_score

fig = plt.figure()
fig.set_size_inches(18.5, 10.5)
score, permutation_scores, pvalue = permutation_test_score(linearmodel, X_train, Y_train, n_permutations=500, n_jobs=1,scoring='neg_mean_squared_error')
plt.hist(permutation_scores, 50, label='Permutation scores',edgecolor='black')
ylim = plt.ylim()
plt.plot(2*[score], ylim, '--g', linewidth=3,label='Classification Score (pvalue %s)' % pvalue)
plt.ylim(ylim)
plt.legend()
plt.xlabel('Score')
plt.show()
```

As seen from above graph, our model works well. As our pvalue<0.005, we can reject the null hypothesis which is that our model is as good as random prediction. So we can say that our linear model is a sound model and does a fairly well job of predicting the logerror. Now let's submit this model's result to zillow competition on kaggle.
